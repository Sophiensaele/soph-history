```python
import matplotlib.pyplot as plt

# Daten vorbereiten
jahre = [1904, 1910, 1920, 1933, 1945, 1961, 1989, 1996, 2000, 2007, 2011, 2023, 1978]

# Zeitstrahl erstellen
fig, ax = plt.subplots(figsize=(12, 2))
ax.eventplot(jahre, orientation='horizontal', colors='black', linelengths=0.4, lineoffsets=0.1)
ax.set_xlim(1900, 2024)
ax.set_ylim(0, 1)

# Jahreszahlen in diagonaler Schrift hinzuf√ºgen, ohne Datenpunkte
for i, jahr in enumerate(jahre):
    ax.text(jahr, 0.5, str(jahr), ha='center', va='center', rotation=45, fontsize=10, color='black')

# Hintergrundfarbe, Titel und Achseneinstellungen
ax.set_facecolor('#cccccc')
ax.yaxis.set_visible(False)
ax.spines[['left', 'top', 'right', 'bottom']].set_visible(False)
ax.xaxis.set_ticks_position('top')
plt.title("Timeline Sophiensaele 1904 - 2024", color='black', fontsize=14, fontweight='bold')
plt.tight_layout()

# Speichern des Zeitstrahls
plt.savefig('timeline_chart_improved_v4.png')
```

```python
import matplotlib.pyplot as plt

# Daten f√ºr die Legende vorbereiten
jahre = [1904, 1910, 1920, 1933, 1945, 1961, 1989, 1996, 2000, 2007, 2011, 2023, 1978]
ereignisse = [
    'Bau des Handwerkervereinshauses',
    'Nutzung durch revolution√§re Linke',
    'Allgemeine politische Versammlungen',
    'NS-Zeit: Nutzung f√ºr Zwangsarbeit',
    'Ende des Zweiten Weltkriegs',
    'Bau der Berliner Mauer',
    'Fall der Berliner Mauer',
    'Gr√ºndung der Sophiens√¶le',
    'Leitung: Amelie Deuflhard',
    'Leitung: Heike Albrecht',
    'Leitung: Franziska Werner',
    'Leitung: Jens Hillje und Andrea Niederbuchner',
    'Werkst√§tten des Maxim-Gorki-Theaters'
]

# Legende chronologisch sortieren
sorted_legend_data = sorted(zip(jahre, ereignisse), key=lambda x: x[0])

# Legende grafisch darstellen
fig, ax = plt.subplots(figsize=(5, 4))
for i, (jahr, ereignis) in enumerate(sorted_legend_data):
    ax.text(0.05, i, f"{jahr}: {ereignis}", verticalalignment='top', fontsize=8)

# Einstellungen f√ºr das Layout der Legende
ax.set_xlim(0, 1)
ax.set_ylim(0, len(ereignisse))
ax.axis('off')
plt.title("Ereignis√ºbersicht", fontsize=12, fontweight='bold')
plt.tight_layout()

# Speichern der Legende als Datei
plt.savefig('legend_chart_improved_v2.png')
```
